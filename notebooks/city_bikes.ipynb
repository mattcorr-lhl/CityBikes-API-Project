# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import requests
import pandas as pd

# URL to be requested
request_url = "http://api.citybik.es/v2/networks/nextbike-glasgow"

# Perform the API call
response = requests.get(request_url)

# 200 = Success; 404 = Not Found
print(response)

# Display the raw JSON data
# print(response.json()['network']) - Commented out, output too long for GitHub
```

Put your parsed results into a DataFrame.

```python
# NOTE: I re-arranged this and the next step. Unsure if that was an acceptable way of doing it

# Load the station data from the JSON response into the pandas dataframe
all_station_data = pd.json_normalize(response.json()['network']['stations'])

all_station_data
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
# Rename a few columns to make them less ambigious
all_station_data.rename(columns={'id':'station_id'}, inplace=True)
all_station_data.rename(columns={'name':'station_name'}, inplace=True)

# Copy the chosen columns into a new dataframe
selected_station_data = all_station_data[['station_id','station_name','free_bikes','empty_slots','latitude','longitude']].copy()

selected_station_data
```

```python
# Export the data into a CSV for use in other notebooks
selected_station_data.to_csv('../data/citybike_stations.csv', index=False)
```
